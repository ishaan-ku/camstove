```python
!pip install pillow-heif
from PIL import Image
from pillow_heif import register_heif_opener
import os # Moved import os to the top

register_heif_opener()

# Replace with the actual path to your photo in Google Drive
heic_input_path = '/content/drive/MyDrive/stove pictures/IMG_5560.HEIC'
png_output_path = '/content/drive/MyDrive/stove/photo.png'

# Create the directory if it doesn't exist
output_directory = os.path.dirname(png_output_path)
os.makedirs(output_directory, exist_ok=True)

img = Image.open(heic_input_path)
img.save(png_output_path, format='PNG') # This line should now work

print(f"Converted single image: {heic_input_path} to {png_output_path}")

# The rest of the code for batch conversion
folder_path = '/content/drive/MyDrive/stove pictures'

for filename in os.listdir(folder_path):
    if filename.lower().endswith(".heic"):
        heic_path = os.path.join(folder_path, filename)
        # Change extension to .png
        png_path = os.path.join(folder_path, filename.rsplit('.', 1)[0] + '.png')

        img = Image.open(heic_path)
        img.save(png_path, format='PNG')
        print(f"Converted: {filename}")
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

folder_path = '/content/drive/MyDrive/stove pictures'
images = [f for f in os.listdir(folder_path) if f.lower().endswith('.png')] # Changed to only look for .png files

# Define grid size (e.g., 2 rows, 5 columns)
rows, cols = 2, 5
plt.figure(figsize=(15, 6))

for i, img_name in enumerate(images[:rows*cols]):
    plt.subplot(rows, cols, i + 1)
    img = mpimg.imread(os.path.join(folder_path, img_name))
    plt.imshow(img)
    plt.axis('off')  # Hide X/Y axes for a cleaner look
    plt.title(img_name, fontsize=8)

plt.tight_layout()
plt.show()
```
