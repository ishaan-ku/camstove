```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

folder_path = '/content/drive/MyDrive/stove pictures'
images = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png'))]

# Define grid size (e.g., 2 rows, 5 columns)
rows, cols = 3, 8
plt.figure(figsize=(15, 6))

for i, img_name in enumerate(images[:rows*cols]):
    plt.subplot(rows, cols, i + 1)
    img = mpimg.imread(os.path.join(folder_path, img_name))
    plt.imshow(img)
    plt.axis('off')  # Hide X/Y axes for a cleaner look
    plt.title(img_name, fontsize=8)

plt.tight_layout()
plt.show()
```
